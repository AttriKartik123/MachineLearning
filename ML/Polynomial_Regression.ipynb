```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


#Importing the dataset
dataset = pd.read_csv('Salary_Positions.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:,2].values
print (X)
print (y)

#Fitting linear regression to dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)

#Fitting polynomial regression to dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)

#visualisiong the Linear Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Linear Regression')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()

#visualisiong the Polynomial Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Polynomial Regression')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()


#visualisiong the Polynomial Regression for smoother curve results
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Polynomial Regression')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()

```

#What is Polynomial Regression?

Polynomial Regression extends beyond straight lines, allowing for curves in data modeling. It's like fitting a curve to your data points, making it a versatile tool for capturing intricate relationships.

#Why Polynomial?

Imagine predicting the price of a product over time. Linear models might fall short when faced with complex price fluctuations. Polynomial Regression, on the other hand, can gracefully adapt to these curves, offering a more realistic prediction.

#How Does it Work?

Polynomial Regression introduces higher-degree terms to the equation, enabling the model to capture intricate patterns in the data. This adaptability makes it a handy tool for scenarios where relationships aren't strictly linear.

Key Considerations:


#Degree of the Polynomial:
 The degree determines the complexity of the curve. A higher degree can lead to overfitting, capturing noise instead of the actual pattern.

#Balancing Complexity:
 It's a trade-off between capturing intricate patterns and avoiding unnecessary complexity. Finding the right balance is crucial.

#Real-world Application:

Polynomial Regression finds applications in various fields. Whether predicting stock prices, analyzing trends in climate data, or understanding customer behavior, its ability to model non-linear relationships makes it a valuable tool.

#Linear Boundaries:

Polynomial Regression, with its simple yet powerful concept, extends the boundaries of linear models. It's a versatile addition to the machine learning toolkit, offering a practical approach to handling diverse datasets.

In our exploration of Polynomial Regression, the key is understanding the data's nuances. This understanding lets us create predictions that better match the real complexities in our datasets. In a nutshell, Polynomial Regression is a versatile tool that helps model complex data relationships more effectively. Its simplicity and adaptability go beyond traditional linear models, making it a valuable asset in the world of machine learning.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Generate synthetic data
np.random.seed(0)
X = np.random.rand(100, 1)
y = 2 * X + 1 + np.random.randn(100, 1)

# Apply Polynomial Regression
degree = 2  # Set the polynomial degree
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X)

# Fit a linear regression model to the polynomial features
model = LinearRegression()
model.fit(X_poly, y)

# Predict the values for the given data
X_new = np.linspace(0, 1, 100).reshape(-1, 1)
X_new_poly = poly_features.transform(X_new)
y_new = model.predict(X_new_poly)

# Plot the data and the polynomial regression curve
plt.scatter(X, y, label='Data', color='blue')
plt.plot(X_new, y_new, label='Polynomial Regression', color='red', linewidth=2)
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title(f'Polynomial Regression (Degree {degree})')
plt.show()
```
